```python
import gdown
import pandas as pd

# Reading the csv file with the gdrive ids
gdrive_ids = pd.read_csv('gdrive_ids.csv')
display(gdrive_ids)
```

```python
### Spoiler raw dataset
## Downloading zip files specific for spoiler detection

## Creating the url
# url for raw spoiler dataset
url = f'https://drive.google.com/uc?id={gdrive_ids[gdrive_ids["name"].str.contains("spoilers_raw.json")]["id"].values[0]}'

# Downloading the file
gdown.download(url, output=None, quiet=False)

# saving the contents of the zipped file
data = pd.read_json("goodreads_reviews_spoiler_raw.json.gz", lines=True, chunksize=100000)
# saving to csv
for i, chunk in enumerate(data):
    chunk.to_csv(f"books/goodreads_reviews_spoiler_raw_{i}.csv", index=False)
```

```python
### Spoiler dataset
## Downloading zip files specific for spoiler detection

## Creating the url
# url for raw spoiler dataset
url = f'https://drive.google.com/uc?id={gdrive_ids[gdrive_ids["name"].str.contains("spoilers.json")]["id"].values[0]}'

# Downloading the file
gdown.download(url, output=None, quiet=False)

# saving the contents of the zipped file
data = pd.read_json("goodreads_reviews_spoiler.json.gz", lines=True, chunksize=100000)
# saving to csv
for i, chunk in enumerate(data):
    chunk.to_csv(f"books/goodreads_reviews_spoiler_{i}.csv", index=False)
```

```python
### Books dataset
## Downloading zip files specific for spoiler detection

## Creating the url
# url for raw spoiler dataset
url = f'https://drive.google.com/uc?id={gdrive_ids[gdrive_ids["name"].str.contains("books.json")]["id"].values[0]}'

# Downloading the file
gdown.download(url, output=None, quiet=False)

# saving the contents of the zipped file
data = pd.read_json("goodreads_books.json.gz", lines=True, chunksize=100000)
# saving to csv
for i, chunk in enumerate(data):
    chunk.to_csv(f"books/goodreads_books_{i}.csv", index=False)
```
